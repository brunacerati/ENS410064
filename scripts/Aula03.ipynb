### 1. Numpy arrays

```python
import numpy as np

#Criando um vetor de dados aleatórios
data = np.random.rand(1000,2) #1000 linhas e 2 colunas

print(data.shape) #verifica o tamanho do vetor/matriz (linhas e colunas)
print(np.size(data)) #quantos dados eu tenho, quantos valores tenho na minha série
print(data[0:10]) #printa os valores da posição 0 a 10
```

```python
# Usando Monte-Carlo
dataMax = 1000
dataMin = 10
data= (dataMax - dataMin)*data + dataMin
print(data[0:10])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3) #cria uma figura/gráfico, aqui nesse caso como tem o 3 dentro de parênteses, vai ser três gráficos
ax[0].plot(data) #plota o gráfico no ax 0 
ax[1].plot(data[:,0])
ax[2].plot(data[:,1])
print(type(data))
```

```python
import pandas as pd

path = 'C:/Users/bruna/Documents/ENS410064/ENS410064/dados/brutos'
#def faz função
def noiseCreation(path,year): 

    #Criando um dataframe
    df = pd.DataFrame()
    
    #criando um vetor temporal
    dtime = np.arange(np.datetime64(str(year)+"-01-01"), np.datetime64(str(year)+"-02-01"), np.timedelta64(1,"h"))
    
    df['datetime'] = dtime
    df = df.set_index(df['datetime']) #setar um índice, antes era 0, 1, 2 ... agora o índice é o datetime
    
    #Criando colunas de ano, mês e dia
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    tt = np.arange(0,df.shape[0],1) #vai criar um vetor de  0 até df shape com espaçamento de 1
    
    #Cria uma coluna random
    df['random'] = np.random.rand(df.shape[0]) 
    
    #Cria sazonalidade
    df['sazonal'] = np.sin(tt/(360*4))+1/2
    df['timeSeries'] = df['sazonal']+df['random']
    df['sazonal'].plot()
    df['timeSeries'].plot(color='green')
    
    #salvando 
    df.to_csv(path+'/noise_'+str(year)+'.csv')

    return df
```
